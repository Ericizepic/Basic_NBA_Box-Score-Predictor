```python
import numpy as np
import tensorflow as tf
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
np.set_printoptions(precision=3, suppress=True)
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler


from tensorflow import keras
from tensorflow.keras import layers
```

```python
dataset = pd.read_csv("C:\\Users\\ericd\\OneDrive\\Documents\\2017games.csv", skiprows = 1, names = ["teamLoc","teamPTS", "teamAST", "teamTO", "teamSTL", "teamBLK", "teamPF","teamFGA", "teamFGM", "teamFG%", "team2PA", "team2PM", "team2P%", "team3PA", 'team3PM', "team3P%", "teamFTA", "teamFTM", "teamFT%", "teamORB", "teamDRB", "teamPTS1","teamPTS2", "teamRslt"])
dataset.head()
```

```python
features = dataset.copy()
labels = features.pop('teamRslt')
```

```python
labels
```

```python
features = np.array(features)
features
```

```python
model = Sequential([
    Dense(units = 32, activation = 'relu'),
    Dense(units= 32, activation = 'relu'),
    Dense(units = 2, activation = None)
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(features, labels, batch_size=10, epochs=10, verbose = 2)
```

```python
testing = pd.read_csv("C:\\Users\\ericd\\OneDrive\\Documents\\2017games.csv", skiprows = 1, names = ["teamLoc","teamPTS", "teamAST", "teamTO", "teamSTL", "teamBLK", "teamPF","teamFGA", "teamFGM", "teamFG%", "team2PA", "team2PM", "team2P%", "team3PA", 'team3PM', "team3P%", "teamFTA", "teamFTM", "teamFT%", "teamORB", "teamDRB", "teamPTS1","teamPTS2", "teamRslt"])
testing.head()
```

```python
testing_features = testing.copy()
testing_label = testing_features.pop('teamRslt')
```

```python
testing_features = np.array(testing_features)
testing_features.shape
```

```python
predictions = model.predict(testing_features, batch_size=10, verbose = 0)
```

```python
for i in predictions:
    print(i)
```

```python
rounded_predictions = np.argmax(predictions,axis=-1)
```

```python
for i in rounded_predictions:
    print(i)
```

```python
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true = testing_label, y_pred = rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes, normalize =False, title = 'Confusion Matrix', cmap = plt.cm.Blues):
    plt.imshow(cm,interpolation = 'nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    
    if normalize:
        cm = cm.astype('float')/cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('confusion matrix, without normlaization')

    print(cm)
    thresh = cm.max()/2.
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i,cm[i,j],
            horizontalalignment = "center",
            color = "white" if cm[i,j] > thresh else "black")
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
        
```

```python
cm_plot_labels = ['no_side_Effects', 'had_side_Effects']
plot_confusion_matrix(cm = cm, classes = cm_plot_labels, title = 'Confusion Matrix')
```
